### Vérifier dataSet

```python
from huggingface_hub import snapshot_download
from pathlib import Path

dataset_dir = snapshot_download(
    repo_id="medbridge-ai/asr-ewe-omni",
    repo_type="dataset"
)

print(dataset_dir)
print(list(Path(dataset_dir).iterdir()))
```

```python
from pathlib import Path

parquets = list(Path(dataset_dir).rglob("*.parquet"))
print("Nombre de fichiers parquet:", len(parquets))
print(parquets[:5])
```

```python
import pandas as pd

df = pd.read_parquet(parquets[0])
print(df.shape)
print(df.columns)
df.head()
```

### Vérifier modèle

```python
model_dir = snapshot_download(
    repo_id="medbridge-ai/omni-ewe-asr",
    repo_type="model"
)

print(model_dir)
print(list(Path(model_dir).iterdir()))
```

```python
expected = [
    "asset_card.yaml",
    "model.yaml",
    "train_config.yaml",
    "model/pp_00/tp_00/sdp_00.pt",
]

for f in expected:
    path = Path(model_dir) / f
    print(f, "OK" if path.exists() else "MISSING")
```

```python
ckpt = Path(model_dir) / "model/pp_00/tp_00/sdp_00.pt"
print("Checkpoint size GB:", ckpt.stat().st_size / 1e9)
```

```python
assert len(parquets) > 0, "Aucun parquet trouvé"
assert (Path(model_dir) / "asset_card.yaml").exists()
assert (Path(model_dir) / "model/pp_00/tp_00/sdp_00.pt").exists()

print("Dataset et modèle OMI Ewe chargés depuis Hugging Face avec succès.")
```
